# Kmeans for dataset1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score

# Load the dataset
data = pd.read_csv(r'D:\OMSCS\Machine learning_Fall_2023\HW3\dataset-1.csv')

# Separate features and the target label
X = data.drop('Target', axis=1)
y_true = data['Target']

# Range of clusters to try
range_n_clusters = range(2, 11)

# Lists to store metrics
wcss = []
silhouette_avg = []
homogeneity_scores = []
completeness_scores = []

for n_clusters in range_n_clusters:
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=10).fit(X)
    cluster_labels = kmeans.predict(X)

    # Calculate metrics
    wcss.append(kmeans.inertia_)
    silhouette_avg.append(silhouette_score(X, cluster_labels))
    homogeneity_scores.append(homogeneity_score(y_true, cluster_labels))
    completeness_scores.append(completeness_score(y_true, cluster_labels))

# Plotting the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.plot(range_n_clusters, wcss, marker='o')
plt.title('WCSS / Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')

plt.subplot(1, 4, 2)
plt.plot(range_n_clusters, silhouette_avg, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')

plt.subplot(1, 4, 3)
plt.plot(range_n_clusters, homogeneity_scores, marker='o')
plt.title('Homogeneity Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Homogeneity Score')

plt.subplot(1, 4, 4)
plt.plot(range_n_clusters, completeness_scores, marker='o')
plt.title('Completeness Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Completeness Score')

plt.tight_layout()
plt.show()
```

# Dimensioality reduction fro dataset 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA
from sklearn.random_projection import GaussianRandomProjection
from sklearn.manifold import MDS 
from scipy.stats import kurtosis
from sklearn.preprocessing import StandardScaler  
from sklearn.metrics import pairwise_distances

# Load the dataset
data = pd.read_csv(r'D:\OMSCS\Machine learning_Fall_2023\HW3\dataset-1.csv')
X = data.drop('Target', axis=1)
y = data['Target']

# Standardizing the data
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)  

# Set up the matplotlib figure
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# PCA - Eigenvalues and Explained Variance
pca = PCA().fit(X_standardized)
axs[0, 0].plot(np.cumsum(pca.explained_variance_ratio_))
axs[0, 0].set_xlabel('Number of Components')
axs[0, 0].set_ylabel('Cumulative Explained Variance')
axs[0, 0].set_title('Explained Variance by Different Principal Components')
axs[0, 0].grid(True)

# ICA - Kurtosis
ica = FastICA(n_components=10, random_state=0)
X_ica = ica.fit_transform(X_standardized)
kurtosis_vals = kurtosis(X_ica, axis=0, fisher=False)
axs[0, 1].plot(range(1, len(kurtosis_vals) + 1), kurtosis_vals, marker='o')  
axs[0, 1].set_xlabel('Independent Components')
axs[0, 1].set_ylabel('Kurtosis')
axs[0, 1].set_title('Kurtosis of Independent Components')
axs[0, 1].grid(True)


# Random Projection - Reconstruction Error
errors = []
for n_components in range(1, X.shape[1]):
    rp = GaussianRandomProjection(n_components=n_components)
    X_rp = rp.fit_transform(X_standardized)
    X_reconstructed = np.dot(X_rp, np.linalg.pinv(rp.components_.T))
    errors.append(np.mean((X_standardized - X_reconstructed) ** 2))

axs[1, 0].plot(range(1, X.shape[1]), errors)
axs[1, 0].set_xlabel('Number of Components')
axs[1, 0].set_ylabel('Reconstruction Error')
axs[1, 0].set_title('Reconstruction Error by Number of Components')

# Compute pairwise dissimilarities (euclidean distances) for your dataset
dissimilarities = pairwise_distances(X_standardized)

# Calculate stress values for different numbers of components
stress_values = []
for n_components in range(1, X.shape[1]):
    mds = MDS(n_components=n_components, random_state=0,max_iter=3, dissimilarity='euclidean')
    mds.fit(dissimilarities)
    stress_values.append(mds.stress_)

# Plot stress values in the existing subplot
axs[1, 1].plot(range(1, X.shape[1]), stress_values, marker='o')
axs[1, 1].set_xlabel('Number of Components')
axs[1, 1].set_ylabel('Stress')
axs[1, 1].set_title('Stress by Number of Components')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r'D:\OMSCS\Machine learning_Fall_2023\HW3\dataset-1.csv')

# Separate features and the target label
X = data.drop('Target', axis=1)
y_true = data['Target']

# Define the range of clusters to try
cluster_range = range(2, 11)

# Lists to keep track of accuracies for each number of clusters
accuracies_gmm = []
accuracies_kmeans = []

# Loop over the range of cluster numbers
for n_clusters in cluster_range:
    # GMM
    gmm = GaussianMixture(n_components=n_clusters, random_state=10).fit(X)
    cluster_labels_gmm = gmm.predict(X)
    X_with_clusters_gmm = np.c_[X, cluster_labels_gmm]

    # K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=10).fit(X)
    cluster_labels_kmeans = kmeans.predict(X)
    X_with_clusters_kmeans = np.c_[X, cluster_labels_kmeans]
    
    # Split the data into training and testing sets for GMM
    X_train_gmm, X_test_gmm, y_train_gmm, y_test_gmm = train_test_split(X_with_clusters_gmm, y_true, test_size=0.2, random_state=42)
    
    # Split the data into training and testing sets for K-Means
    X_train_kmeans, X_test_kmeans, y_train_kmeans, y_test_kmeans = train_test_split(X_with_clusters_kmeans, y_true, test_size=0.2, random_state=42)
    
    # Create an instance of MLPClassifier and fit the data for GMM
    mlp_gmm = MLPClassifier(random_state=42)
    mlp_gmm.fit(X_train_gmm, y_train_gmm)
    
    # Create an instance of MLPClassifier and fit the data for K-Means
    mlp_kmeans = MLPClassifier(random_state=42)
    mlp_kmeans.fit(X_train_kmeans, y_train_kmeans)
    
    # Make predictions and evaluate for GMM
    y_pred_gmm = mlp_gmm.predict(X_test_gmm)
    accuracies_gmm.append(accuracy_score(y_test_gmm, y_pred_gmm))
    
    # Make predictions and evaluate for K-Means
    y_pred_kmeans = mlp_kmeans.predict(X_test_kmeans)
    accuracies_kmeans.append(accuracy_score(y_test_kmeans, y_pred_kmeans))

# Create the plot with a specified figure size
plt.figure(figsize=(7, 1))
plt.plot(cluster_range, accuracies_gmm, marker='o', label='GMM')
plt.plot(cluster_range, accuracies_kmeans, marker='x', label='K-Means')
plt.title('ANN Accuracy vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
